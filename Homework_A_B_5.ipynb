```python
import pandas as pd
pd.options.mode.chained_assignment = None  # default='warn'
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats
import scipy
from pylab import rcParams
rcParams['figure.figsize'] = 13,6
data = pd.read_csv('Netology_A_B-testing_HW_1.csv')
pd.options.display.max_rows = 1000
```

```python
data_city_center_order = data.loc[data['city_center_order'] == 1]
data_city_center_order
```

```python
half = len(data_city_center_order['user_cancelled']) / 2
sam1 = data_city_center_order.loc[:half, 'user_cancelled']
sam2 = data_city_center_order.loc[half:, 'user_cancelled']
scipy.stats.ttest_ind(sam2, sam1)
```

```python
dfs = (half - 1) + (half - 1)
scipy.stats.t.ppf(0.975, dfs)
```
