```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
import json
```

```python
keywords = ['python', 'парсинг']
res = requests.get('https://habr.com/ru/all/')
soup = BeautifulSoup(res.text, 'html.parser')
```

```python
posts = soup.find_all('article', class_='post') #выделяю пост
```

```python
prew = soup.find_all('div', class_ = 'post__text') #Выделяю превью
```

```python
data_1 = []
for post in posts:
    hubs = post.find_all('div', class_ = 'post__text')
    for hub in hubs:
        hub_lower = hub.text.lower()
        if ([hub_lower in key for key in keywords]):
            title_element = post.find('a', class_='post__title_link')
            time_element = post.find('span', class_='post__time')
        data = {'date' : time_element.text, 'title': title_element.text, 'link': title_element.attrs.get('href')}
        data_1.append(data)
        #print(time_element.text, title_element.text, title_element.attrs.get('href'))
```

```python
frame = pd.DataFrame(data_1)
frame
```

```python
url = 'https://identityprotection.avast.com/v1/web/query/site-breaches/unauthorized-data'
params = {'emailAddresses': ['xxx@mail.ru']}
headers_ = {
    'Vaar-Version': '0',
    'Vaar-Header-App-Product': 'hackcheck-web-avast'
}
res = requests.post(url, json = params, headers = headers_)
res
```

```python
data_2 = []
breach = res.json()['breaches']
for key, value in breach.items():
    data = {'email': params['emailAddresses'], 'date' : value['publishDate'], 'sourse': value['site'], 
            'description': value['description']}
    data_2.append(data)
```
