```python
import pandas as pd
pd.options.mode.chained_assignment = None  # default='warn'
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats
import scipy
from pylab import rcParams
rcParams['figure.figsize'] = 13,6
data = pd.read_csv('Netology_A_B-testing_HW_1.csv')
pd.options.display.max_rows = 1000
```

```python
data
```

Заказ из центра => более низкие юзерские отмены

```python
data_city_center_order = data.loc[data['city_center_order'] == 1]
data_city_center_order
```

```python
half = len(data_city_center_order['order_made']) / 2
sam1 = data_city_center_order.loc[:half, 'order_made']
sam2 = data_city_center_order.loc[half:, 'order_made']
scipy.stats.ttest_ind(sam2, sam1)
```

```python
dfs = (half - 1) + (half - 1)
scipy.stats.t.ppf(0.975, dfs)
```

```python
"""
pvalue 0,82 больше 0,05, значит средние двух выборок равны.
табличное значение 1,96 больше statistic 0,22, значит средние двух выборок равны при условии их нормального распределения
"""
```
