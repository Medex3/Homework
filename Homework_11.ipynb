```python
import pandas as pd
ratings = pd.read_csv('ratings.csv')
keywords = pd.read_csv('keywords.csv')
movies = pd.read_csv('movies.csv')
movies_rating = pd.merge(movies, ratings, on = 'movieId') #Объединенная таблица из movies и ratings по ID
```

```python
def rate(param):
    if param <= 2:
        return 'Низкий'
    elif 2 < param <= 4:
        return 'Средний'
    else:
        return 'Высокий'
```

```python
ratings['class'] = ratings['rating'].apply(rate) 
ratings.head(200)
#Формирую столбец class, в кот. записывается значение согласно функции. rating идет в param функции
```

```python
geo_data = {
'Центр': ['москв', 'тула', 'ярославль'],

'Северо-Запад': ['петербург', 'псков', 'мурманск'],

'Дальний Восток': ['владивосток', 'сахалин', 'хабаровск']}
```

```python
def region(param):
    for key, value in geo_data.items(): #Разбиваю словарь на ключи и значения
        for i in value: #Для каждого элемента в значениях проверяю наличие в столбце keyword
            if i in param['keyword']:
                return key
            else:
                return 'undefined'
```

```python
keywords['region'] = keywords.apply(region, axis = 1)
keywords.head(200)
```

```python
years = list(range(1950, 2011))
import re
```

```python
def production_year(row):
    """
    Присваивает каждоу фильму год, если он есть
    """
    for i in row:
        array = re.findall(r'\d{4}', row) #Регулярное выражение для определения 4 последовательно записанных цифр
        if len(array) == 0: #Проверка, если нет чисел в строке
            array = [1900] 
        elif int(array[0]) not in years: #Беру первый(и единственный) элемент списка array, перевожу в формат int
            array = [1900]
        s = pd.Series(array) #Список array перевожу в формат Series
        return s
```

```python
movies_rating['years'] = movies_rating['title'].apply(production_year)
#print(movies_rating.groupby('movieId'))
#movies_rating
```

```python
movies_rating.groupby('years').agg({'rating': 'mean'}).sort_values('rating', ascending = False)
```
