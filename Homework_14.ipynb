```python
import pandas as pd
import numpy as np
import re
name = ['surgery', 'Age', 'Hospital Number', 'rectal_temperature', 'pulse', 'respiratory_rate', 'temperature_of_extremities', 
        'peripheral_pulse', 'mucous_membranes', 
        'Source1', 'Past Usage1', 'Relevant1', 'Instances1', 'attributes1', '11', '22',
       'TItle2', 'Source2', 'Past Usage2', 'Relevant2', 'Instances2', 'attributes2', '12', '223',
       'TItle3', 'Source3', 'Past Usage3', 'Relevant3']
data = pd.read_csv('horse_data.csv', names = name, na_values='?') #Чтение файла с использованием названий столбцов
data_sort = data.loc[:, 'surgery': 'mucous_membranes'] #Вывод нужных 8 столбцов
#data_sort_fill = data.loc[:, 'surgery': 'mucous_membranes'] #Отдельный файл для заполнения пропусков

#data_sort = data_sort.replace(r'\?', np.nan, regex = True) #Замена '?' на NaN
```

```python
for col in data_sort.columns:
    pct_missing = data_sort[col].isna().mean()
    print(f'{col} - {pct_missing :.1%}')
```

```python
data_sort.describe()
```

```python
for key, value in data_sort.items():
    print('Значение моды: ', value.round().mode()[0])
```

```python
q1 = data_sort.surgery.quantile(0.25)
q3 = data_sort.surgery.quantile(0.75)
iqr = q3 - q1
lower_bound = q1 - (1.5 * iqr) 
upper_bound = q3 + (1.5 * iqr)
remove_outliers = data_sort[data_sort.surgery.between(lower_bound, upper_bound, inclusive = True)]
pd.concat([data_sort, remove_outliers]).drop_duplicates(keep = False)

#Единственный выброс - Nan. Заполню значением моды
```

```python
data_sort.surgery.replace(np.NaN, data_sort.surgery.round().mode()[0], inplace = True) #Заменяю пропус модой
```

```python
q1 = data_sort.Age.quantile(0.25)
q3 = data_sort.Age.quantile(0.75)
iqr = q3 - q1
lower_bound = q1 - (1.5 * iqr) 
upper_bound = q3 + (1.5 * iqr)
remove_outliers = data_sort[data_sort.Age.between(lower_bound, upper_bound, inclusive = True)]
pd.concat([data_sort, remove_outliers]).drop_duplicates(keep = False)

#Выбросы больше допустимых значений. Заполню значением "2", т.к. это значение прописано в описании, но нигде не фигурирует,
#поэтому, возможно, вместо 2 указано 9
```

```python
data_sort.Age = data_sort.Age.replace(9, 2) #Заменяю 9 на 2
```

```python
q1 = data_sort.rectal_temperature.quantile(0.25)
q3 = data_sort.rectal_temperature.quantile(0.75)
iqr = q3 - q1
lower_bound = q1 - (1.5 * iqr) 
upper_bound = q3 + (1.5 * iqr)
remove_outliers = data_sort[data_sort.rectal_temperature.between(lower_bound, upper_bound, inclusive = True)]
pd.concat([data_sort, remove_outliers]).drop_duplicates(keep = False)
#print(q1, q3)
#Нормальная температура лошади считается 37,2 - 38,5
#Выбросы меньше 37,2 заменяю на моду (меньше средней температуры), выше 38,5 оставляю, т.к., скорее всего, говорят о болезни
```

```python
for i in data_sort.rectal_temperature: #Заменяю модой выбросы
    if i < 37.2:
        data_sort.rectal_temperature = data_sort.rectal_temperature.replace(i, data_sort.rectal_temperature.round().mode()[0])
```

```python
data_sort.rectal_temperature.replace(np.NaN, data_sort.rectal_temperature.median(), inplace = True) #Заполнил пропуски медианой
```
