```python
import pandas as pd
data = pd.read_csv('ratings.csv') #Сформировал таблицу из файла
```

```python
data_rating = data[ data['rating'] == 5.0]  #Сформировал отдельную таблицу, где указаны только строки с оценкой 5
film = data_rating['movieId'].value_counts().head(1)
print('Фильм с ID', film.index.tolist(), 'имеет наибольшее количество оценок "5" - ', film.tolist())
```
